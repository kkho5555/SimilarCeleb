```python
import cv2
import sys
import os
```

```python
# 입력 파일 지정하기
image_file = "train\\"
# 캐스케이드 파일의 경로 지정하기 --- (※1)
cascade_file = "haarcascade_frontalface_alt.xml"
# 이미지 읽어 들이기 --- (※2)
for i in range(1,10):
    print(len(os.listdir(image_file+str(i))))
```

```python
for i in range(10,13):
    for idx, img in enumerate(os.listdir(image_file+str(i))):
        if idx < 10:
            idx = "000"+str(idx)
        elif idx <100:
            idx = "00"+str(idx)
        elif idx < 1000:
            idx = "0"+str(idx)
        os.rename(image_file+str(i)+"/"+img, image_file+str(i)+"/"+f"{i}_{idx}.jpg")
```

```python
for i in range(1,10):
    for img in os.listdir(image_file+str(i)):        
        image = cv2.imread(image_file+str(i)+'/'+img, cv2.IMREAD_COLOR)
        # 그레이스케일로 변환하기
        image_gs = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        # 얼굴 인식 특징 파일 읽어 들이기 --- (※3)
        cascade = cv2.CascadeClassifier(cascade_file)
        # 얼굴 인식 실행하기

        faces = cascade.detectMultiScale(image_gs)        
        if len(faces) > 0:
            # 인식한 부분 표시하기 --- (※4)
            for face in faces:
                x,y,w,h = face
                crop = image_gs[y:y+h, x:x+w]                                
                cv2.imwrite(image_file+str(i)+"/"+img, crop)
                print("save success.")
            # 파일로 출력하기 --- (※5)
        else:
            print("no face")
            os.remove(image_file+str(i)+"/"+img)
```
